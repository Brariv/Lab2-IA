## b. Implemente estructuras de colas FIFO, LIFO y PRIORITY con los siguientes métodos:

- EMPTY: retorna TRUE solamente si la cola se encuentra vacía.
- TOP: retorna el primer elemento de la lista.
- POP: retorna el primer elemento de la lista Y lo remueve de la lista.
- ADD: Inserta 1 elemento en la lista y retorna la lista con el elemento en su posición correspondiente.

```python
class FIFO:
    def __init__(self):
        self.items = []

    def EMPTY(self):
        return len(self.items) == 0

    def TOP(self):
        if self.EMPTY():
            return None
        return self.items[0]

    def POP(self):
        if self.EMPTY():
            return None
        return self.items.pop(0)

    def ADD(self, element):
        self.items.append(element)
        return self.items


class LIFO:
    def __init__(self):
        self.items = []

    def EMPTY(self):
        return len(self.items) == 0

    def TOP(self):
        if self.EMPTY():
            return None
        return self.items[-1]

    def POP(self):
        if self.EMPTY():
            return None
        return self.items.pop()

    def ADD(self, element):
        self.items.append(element)
        return self.items


class Priority:
    def __init__(self):
        self.items = []

    def EMPTY(self):
        return len(self.items) == 0

    def TOP(self):
        if self.EMPTY():
            return None
        return self.items[0]

    def shiftDown(self, index):
        left = 2 * index + 1
        right = 2 * index + 2
        largest = index

        if left < len(self.items) and self.items[left] > self.items[largest]:
            largest = left
        if right < len(self.items) and self.items[right] > self.items[largest]:
            largest = right
        if largest != index:
            self.items[index], self.items[largest] = self.items[largest], self.items[index]
            self.shiftDown(largest)
    
    def shiftUp(self, index):
        while index > 0 and self.items[index] > self.items[(index - 1) // 2]:
            parent = (index - 1) // 2
            self.items[index], self.items[parent] = self.items[parent], self.items[index]
            index = parent

    def POP(self):
        size = len(self.items)
        if self.EMPTY():
            return None
        result = self.items[0]
        self.items[0] = self.items[size - 1]
        self.items.pop()
        self.shiftDown(0)
        return result

    def ADD(self, element):
        self.items.append(element)
        self.shiftUp(len(self.items) - 1)
        return self.items
```

### Tree

```python
#  Make tree with excel files readed

class TreeNode:
    def __init__(self, name):
        self.name = name
        self.children = []
        self.heuristic = None
        self.cost = []

    def add_child(self, child_node, cost):
        self.children.append(child_node)
        self.cost.append(cost)

tree = []

# nodes are made from heuristica

for row in range(0, Heuristica.max_row):
    node_name = Heuristica.cell(row=row + 1, column=1).value
    heuristic_value = Heuristica.cell(row=row + 1, column=2).value
    node = TreeNode(node_name)
    node.heuristic = heuristic_value
    tree.append(node)

# Paths are made from costo funciones

for row in range(0, Costo_Funciones.max_row):
    parent_name = Costo_Funciones.cell(row=row + 1, column=1).value
    child_name = Costo_Funciones.cell(row=row + 1, column=2).value
    cost_value = Costo_Funciones.cell(row=row + 1, column=3).value

    parent_node = next((node for node in tree if node.name == parent_name), None)
    child_node = next((node for node in tree if node.name == child_name), None)

    if parent_node and child_node:
        parent_node.add_child(child_node, cost_value)


print("Tree structure:")
for node in tree:
    print(f"Node: {node.name}, Heuristic: {node.heuristic}")
    for child, cost in zip(node.children, node.cost):
        print(f"  Child: {child.name}, Cost: {cost}")
```

### 1. Breadth-first search

```python
def bfs(nodes):
    if not nodes:
        return []

    q = FIFO()
    visited = set()
    res = []

    start = nodes[0]
    visited.add(start.name)
    q.ADD(start)

    while not q.EMPTY():
        curr = q.POP()
        res.append(curr.name)

        for child in curr.children:
            if child.name not in visited:
                visited.add(child.name)
                q.ADD(child)

    return res

bfs_result = bfs(tree)
print("BFS Result:", bfs_result)

```
